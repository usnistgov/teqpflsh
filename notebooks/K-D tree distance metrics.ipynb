```python
import teqpflsh
import numpy as np
import matplotlib.pyplot as plt
```

```python
xpts = np.linspace(0, 1, 10)
ypts = np.linspace(0, 1, 10)
XPTS, YPTS = [_.ravel() for _ in np.meshgrid(xpts, ypts)]
holder = teqpflsh.L2TreeHolder(XPTS, YPTS, 10, None, None)
tree = holder.tree
plt.plot(XPTS, YPTS, 'k.', ms=2)

# A box, randomly sampled
xrng = np.random.random((10000,))
yrng = np.random.random((10000,))

d2 = [tree.get_nearest_indexd2(x_, y_)[1] for x_,y_ in zip(xrng, yrng)]
plt.scatter(xrng, yrng, c=d2, s=5)
cb = plt.colorbar()
cb.set_label('$d^2$')
plt.show()
```

```python
xpts = np.linspace(1, 10, 10)
ypts = np.linspace(1, 2, 10)
XPTS, YPTS = [_.ravel() for _ in np.meshgrid(xpts, ypts)]
xscaler = teqpflsh.MinMaxLogScaler(1, 2)
holder = teqpflsh.L2TreeHolder(XPTS, YPTS, 10, xscaler, None)
tree = holder.tree
plt.plot(XPTS, YPTS, 'k.', ms=2)

# A box, randomly sampled
xrng = np.random.random((10000,))*9 + 1
yrng = np.random.random((10000,)) + 1

d2 = [tree.get_nearest_indexd2(np.array([x_, y_]))[1] for x_,y_ in zip(xrng, yrng)]
plt.scatter(xrng, yrng, c=d2, s=5)
cb = plt.colorbar()
cb.set_label('$d^2$')
plt.xscale('log')
```
